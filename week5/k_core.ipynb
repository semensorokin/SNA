```
import networkx 
```

```
les_miserables = networkx.generators.social.les_miserables_graph()
```

```
import matplotlib.pyplot as plt

def draw_g(les_miserables):
  pos = networkx.spring_layout(les_miserables)
  print(pos)
  plt.figure(figsize=(20,10))

  options = {
      "font_size": 18,
      "node_size": 1500,
      "node_color": "white",
      "edgecolors": "grey",
      "linewidths": 3,
      "width": 3,
      "with_labels": True
  }
  networkx.draw(les_miserables, pos, **options)
  plt.show()


draw_g(les_miserables)
```

```
les_miserables_k2 = networkx.algorithms.core.k_core(les_miserables, k = 2)
```

```
draw_g(les_miserables_k2)
```

```
les_miserables_k3 = networkx.algorithms.core.k_core(les_miserables, k = 3)
```

```
draw_g(les_miserables_k3)
```

```
les_miserables_k4 = networkx.algorithms.core.k_core(les_miserables, k = 4)
```

```
draw_g(les_miserables_k4)
```

```
colors = ["#FF00FF", "#FFD700", "#FF0000"]

node_color = { i: "#B0C4DE" for i in  les_miserables.nodes}
print(node_color)
for i, nodes in  enumerate([les_miserables_k2.nodes, les_miserables_k3.nodes, les_miserables_k4]):
  for n in nodes: 
    node_color[n] = colors[i]
print(node_color)
```

```
pos = networkx.spring_layout(les_miserables)
plt.figure(figsize=(20,10))

options = {
    "font_size": 18,
    "node_size": 1500,
    "node_color": "white",
    "edgecolors": "grey",
    "linewidths": 3,
    "width": 3,
    "with_labels": True, 
    "node_color": node_color.values()
    
}
networkx.draw(les_miserables, pos, **options)
plt.show()
```

```
r = networkx.degree_assortativity_coefficient(les_miserables)
print(f"{r:3.1f}")
r = networkx.degree_assortativity_coefficient(les_miserables_k2)
print(f"{r:3.1f}")
r = networkx.degree_assortativity_coefficient(les_miserables_k3)
print(f"{r:3.1f}")
r = networkx.degree_assortativity_coefficient(les_miserables_k4)
print(f"{r:3.1f}")
```

```
from networkx.utils import cuthill_mckee_ordering
#G = networkx.barabasi_albert_graph(100, 5)
G = networkx.powerlaw_cluster_graph(100, 5, 0.7)
#G = networkx.karate_club_graph()
default_m = networkx.adjacency_matrix(G) 
plt.spy(default_m, markersize=4)
plt.show()
```

```
rcm = list(cuthill_mckee_ordering(G))
reordered_m = networkx.adjacency_matrix(G, nodelist=rcm) 
plt.spy(reordered_m, markersize=4)
plt.show()
```
