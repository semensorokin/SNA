```python
import pandas as pd
```

```python
youngturks = pd.read_csv("3Y8kG48zSs2PJBuPM3rN3Q_a85679338a8848ea9df3d29cf5a90253_theyoungturks_tutorial.csv")
```

```python
youngturks.head(6)
```

```python
prepared_youngturks = pd.DataFrame(
    {
"Source": youngturks["web-scraper-start-url"].tolist() * 2,
"Target": youngturks["featured-href"].tolist() + youngturks["recommended-href"].tolist()
})
prepared_youngturks["Link"] = 1
prepared_youngturks
```

```python
foxnews = pd.read_csv("ElctOx73Tc2XLTse903NyA_7a98205b347641ffbb1ca1e939244936_foxnews_tutorial.csv")
```

```python
prepared_foxnews = pd.DataFrame(
    {
"Source": foxnews["web-scraper-start-url"].tolist() * 2,
"Target": foxnews["featured-href"].tolist() + foxnews["recommended-href"].tolist()
})
prepared_foxnews["Link"] = 1
prepared_foxnews
```

```python
prepared_data = pd.concat([prepared_foxnews, prepared_youngturks])
```

```python
prepared_data.tail()
```

```python
prepared_data = prepared_data.reset_index(drop=True)
```
